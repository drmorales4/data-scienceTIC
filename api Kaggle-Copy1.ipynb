```python
# metodo para convertir en df los metadatos extraidos

def convert_to_df(csv):
    df = pd.DataFrame(data=csv)[0].str.split(',',expand=True)  
    header = df.iloc[0]
    df = df[1:]
    df.columns = header
    return df
```

```python
fashion_minst_search_in_datasets_csv = !kaggle datasets list -s Machine-learning --csv
fashion_minst_search_in_datasets_df = convert_to_df(fashion_minst_search_in_datasets_csv)
print('Search Results for Fashion-MNIST Dataset:')
ICD.display(fashion_minst_search_in_datasets_df.head(10))
```

```python
# extraer competitions

def getCompetitions(topics):
    competitions_list_csv = !kaggle competitions list -s $topics --csv

    competitions_list_df = convert_to_df(competitions_list_csv)
    
    ICD.display(competitions_list_df.head())
    
    return(competitions_list_csv)
```

```python
listCompetitions = []

for x in topics:
    listCompetitions.append(getCompetitions(x))
```

```python
for x in listCompetitions:
    competitions_list_df = convert_to_df(x)
    
    ICD.display(competitions_list_df.head(10))
```

```python
# extraer datasets

def getDatasets(topics):
    datasets_list_csv = !kaggle datasets list -s $topics --csv
    
    datasets_list_df = convert_to_df(datasets_list_csv)
    
    datasets_list_df_sorted = datasets_list_df.sort_values(by='lastUpdated', ascending=0)
    
    ICD.display(datasets_list_df_sorted.head())
    
    return(datasets_list_csv)
```

```python
listDatasets = []

for y in topics:
    listDatasets.append(getDatasets(y))
```
