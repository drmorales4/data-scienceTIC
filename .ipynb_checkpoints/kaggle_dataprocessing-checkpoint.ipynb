```python
import pandas as pd
import numpy as np
import json
import pprint
```

```python
# leer archivo Excel que contiene los topicos de Fundamentos de Analsis de Datos

dataDataSets = pd.read_csv('APIdata/datasets.csv')
```

```python
dataDataSets.head()
```

```python
dataDataSets.shape
```

```python
dataDataSets.info()
```

```python
# revisar nulos
dataDataSets.isnull().sum()
```

```python
# eliminar columnas

dataDataSets = dataDataSets.drop(['Unnamed: 0'], axis=1)
```

```python
# cambiar de nombre a las columnas

dataDataSets = dataDataSets.rename(columns={'set':'type', 'downloadCount':'downloads'})
```

```python
dataDataSets.head()
```

```python
# extraer los autores desde la columna ref

autores = []

for x in dataDataSets['ref']:
    
    getAutor = x.split('/')
    
    autores.append(getAutor[0]) # obtener todos los autores
    
    print(getAutor)
```

```python
autores
```

```python
# agregar autores al dataframe

dataDataSets.insert(3, "autor", autores)
```

```python
dataDataSets.head()
```

```python
# crear la url de kaggle 
url = []

for x in dataDataSets['ref']:
    newUrl = 'kaggle.com/datasets/' + x
    
    url.append(newUrl)
```

```python
url
```

```python
# agregar otra columna, url al dataframe

dataDataSets.insert(3, "urlKaggle", url)
```

```python
dataDataSets.head()
```

```python
# generar csv

dataDataSets.to_csv("dataStorage/datasets_dataStorage.csv")
```

```python
# convertir dataframe a json

jsonDataSets = dataDataSets.to_json(orient = 'records')
```

```python
jsonDataSets
```

```python
# escribir json

with open('dataStorage/jsonDataSets_dataStorage.json', 'w') as file:
    json.dump(jsonDataSets, file, indent=4)
```
