```python
# Importamos PyMongo y creamos una conexión a la base de datos

import pymongo
from pymongo import MongoClient
import pandas as pd
import json
```

# Conexion a MongoDB

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
# Creamos una base de datos y una colección
db = client["itinerario"]
```

```python
# creamos la coleccion donde vamos a guardar los datos

coleccion = db["dataScienceYouTube_TIC"]
```

# Guardar YouTube

```python
# leer csv

data3 = pd.read_csv('../dataProcessing/dataCleansing/dataYoutube_dataStorage.csv')
data3.head()
```

```python
data3 = data3.drop(['Unnamed: 0'], axis=1)
data3.head()
```

```python
data3 = data3.to_dict(orient='records') # To convert into a dictionary structure
print(data3[:3])
```

```python
# Insertamos el documento en la colección

coleccion.insert_many(data3)
```

```python
# Cerramos la conexión a la base de datos
client.close()
```
