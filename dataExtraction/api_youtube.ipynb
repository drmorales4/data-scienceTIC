```python
# obtener los metadatos de las playlists de los canales

def obtenerPlaylists(channelList):
    
    channel_request = youtube.playlists().list(
        part='contentDetails, snippet', channelId=channelList)
    
    # extraer datos
    channel_response = channel_request.execute()
    
    # print(pl_response)
    
    return(channel_response)
```

```python
channelResponse = []

for x in channelList:
    channelResponse.append(obtenerPlaylists(x))

channelResponse
```

```python
playlist = []

for y in channelResponse:
    playlist.append(y['items'])

playlist
```

```python
playlistIDs = []

for z in playlist:
    for a in z:
        IDs = a['id']
        #print(IDs)
        playlistIDs.append(IDs)
        
playlistIDs
```

```python
def obtenerMetadatosPlaylists(playlistIDs):
    playlists_request = youtube.playlistItems().list(part='contentDetails', playlistId=playlistIDs)
    playlists_response = playlists_request.execute()
    #print(play_response)
    
    return(playlists_response)
```

```python
playlistsResponse = []

for c in playlistIDs:
    playlistsResponse.append(obtenerMetadatosPlaylists(c))

playlistsResponse
```

```python
listVideos = []
```

```python
for d in playlistsResponse:
    videos = d['items']
    #print(videos)
    
    for e in videos:
        IDvideos = e['contentDetails']['videoId']
        #print(IDvideos)
        
        listVideos.append(IDvideos)
    
listVideos
```
