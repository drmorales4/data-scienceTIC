```python
# metodo para convertir en df los metadatos extraidos

def convert_to_df(csv):
    df = pd.DataFrame(data=csv)[0].str.split(',',expand=True)  
    header = df.iloc[0]
    df = df[1:]
    df.columns = header
    return df
```

## Get datasets

```python
# extraer competitions

def getDataSets(tema):
    
    # extraccion de datos de la api, devuelve una lista
    new_datasetsList = !kaggle datasets list -s $tema --csv
    
    # lista vacia
    datasetsList = []
    
    # for para reemplazar las comas (,)
    # al trasformar la lista a dataframe toma la coma (,) de los metadatos como un separador
    for x in new_datasetsList:
        reemplazar = x.replace(', ', ' ')
        datasetsList.append(reemplazar)
    
    # print(datasetsList)
    
    # condicional si no encuentra metadatos de 
    if datasetsList[0] != 'No datasets found':
                
        # convertir lista a dataframe
        datasetsDF = convert_to_df(datasetsList)
        
        # agregar 2 columnas al dataframe
        datasetsDF.insert(0,"topic",tema,True) # topic
        datasetsDF.insert(1,"set","datasets",True) # type
        
        ultimaColumna = datasetsDF.iloc[:,-1:] # sacar la ultima columna para verificar si est√° vacia o no
        
        # condicional if si la ultima columna esta vacia
        if ultimaColumna.columns.values[0] == None:
            
            datasetsDF = datasetsDF.drop([None], axis=1) # eliminar columna vacia
        
        datasetsDF.dropna(inplace=True) # eliminar filas nulas
                
        return(datasetsDF)
    
    # si no se encuentran datasets se retorna un dataframe vacio
    return(pd.DataFrame())
```

```python
cont = 0

# dataframe vacio para concatenar
datasets= pd.DataFrame()

for x in topics:
    cont+=1
    
    # obtenemos los dataframes haciendo uso de la funcion
    dfDatasets = getDataSets(x)
    
    # conficional si el dataframe est√° vacio
    if not dfDatasets.empty:
         
        ICD.display(dfDatasets.head()) # imprimir
    
        # concatenamos todos los dataframes  
        datasets = pd.concat([datasets, dfDatasets])

        # generamos csv de los dataframes
        dfDatasets.to_csv("api_dataExtraction/datasets_" + x + str(cont) + ".csv")
```

## Get notebooks/kernels

```python
# extraer notebooks

def getKernels(topics):
    
    # extraccion de datos de la api, devuelve una lista
    new_kernelsList = !kaggle kernels list -s $topics --csv
    
    # si no se obtiene notebooks se muestra un error 500
    lista1 = new_kernelsList[0]
    # obtener el 500 para condicional
    lista1 = lista1.split('-')[0]
    # print(lista1)
    
    # dataframe vacio para concatenar
    kernelsList = []
    
    # for para reemplazar las comas (,)
    # al trasformar la lista a dataframe toma la coma (,) de los metadatos como un separador
    for x in new_kernelsList:
        reemplazarK = x.replace(', ', ' ')
        kernelsList.append(reemplazarK)
    
    # print(kernelsList)
    
    # condicional si se encuentran o no metadatos
    if lista1 != "500 ":
                
        kernelsDF = convert_to_df(kernelsList) # convertir a df
        
        kernelsDF.insert(0,"topic",topics,True) # agregar columna con el tema
        kernelsDF.insert(1,"set","kernels",True) # agregar columnna con el tipo de metadato
        
        ultimaColumna = kernelsDF.iloc[:,-1:] # sacar la ultima columna para verificar si est√° vacia o no
        
        # condicional if si la ultima columna esta vacia
        if ultimaColumna.columns.values[0] == None:
            
            kernelsDF = kernelsDF.drop([None], axis=1) # eliminar columna vacia
        
        kernelsDF.dropna(inplace=True) # eliminar filas nulas
        
        ICD.display(kernelsDF.head()) # imprimir
                    
        return(kernelsDF) # return
    return(pd.DataFrame())
```

```python
count = 0

# dataframe vacio para concatenar dataframes
notebooks = pd.DataFrame()

for x in topics:
    
    # obtener metadatos por medio de la funcion
    dfKernels = getKernels(x)
    
    if not dfKernels.empty:
    
        # unir todos los dataframes
        notebooks = pd.concat([notebooks, dfKernels])
    
        # generar csv individuales
        dfKernels.to_csv("api_dataExtraction/notebook_" + x + str(cont) + ".csv")
```
